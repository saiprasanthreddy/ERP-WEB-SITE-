```python
# Task 2

attendance_days = {
    "Day 1": ("2025-07-28 18:00:00", "2025-07-29 17:59:59"),
    "Day 2": ("2025-07-29 18:00:00", "2025-07-30 17:59:59"),
    "Day 3": ("2025-07-30 18:00:00", "2025-07-31 17:59:59"),
    "Day 4": ("2025-07-31 18:00:00", "2025-08-01 17:59:59"),
    "Day 5": ("2025-08-01 18:00:00", "2025-08-02 17:59:59")
}


df['Unique ID'] = df['Verified Email'].fillna(df['Email Id']).fillna(df['WhatsApp Number'])

attendance = df[['Unique ID', 'Email Id', 'Full Name']].drop_duplicates().set_index('Unique ID')

for day, (start, end) in attendance_days.items():
    mask = (df['Timestamp'] >= pd.to_datetime(start)) & (df['Timestamp'] <= pd.to_datetime(end))
    attendees = df[mask]['Unique ID'].unique()
    attendance[day] = attendance.index.isin(attendees)

attendance = attendance.reset_index()
for day in attendance_days:
    attendance[day] = attendance[day].map({True: '✔️', False: '❌'})

attendance.to_csv("attendance_report.csv", index=False)
```

```python
# Task 3


df_cleaned = pd.read_csv("cleaned_master.csv")

# Report B: College-Wise Participation
college_report = df_cleaned.groupby("College Full Name")["Email Id"].nunique().reset_index(name="No. of Students")
college_report.to_csv("college_wise_participation.csv", index=False)

# Report C: District-Wise Participation
district_report = df_cleaned.groupby("College District")["Email Id"].nunique().reset_index(name="Student Count")
district_report.to_csv("district_wise_participation.csv", index=False)

# Report D: State-Wise Participation
state_report = df_cleaned.groupby("College State")["Email Id"].nunique().reset_index(name="No. of Students")
state_report.to_csv("state_wise_participation.csv", index=False)

# Report E: Top Branches
branch_report = df_cleaned.groupby("Branch Name")["Email Id"].nunique().reset_index(name="Student Count")
branch_report.to_csv("top_branches.csv", index=False)
```
